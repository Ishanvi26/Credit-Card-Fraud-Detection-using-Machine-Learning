# Undersampling

As our dataset is imbalanced, we are going to create a sample dataset containing similar distribution of legal and fraudulent transactions

Number of fraudulent transactions - 492

```python
fraud.head()
```

```python
legit.head()
```

```python
legit = legit.sample(n=492)
```

```python
legit.head()
```

```python
legit.shape, fraud.shape
```

```python
new_ds = pd.concat([legit, fraud], axis=0)
```

```python
new_ds.head()
```

```python
new_ds.shape
```

```python
new_ds["Class"].value_counts()
```

Now we have an equal distribution of both data

```python
new_ds.groupby("Class").mean()
```

```python
new_ds.to_csv('balanced_dataset.csv', index=False, encoding='utf-8')
```

# Splitting into Dependent and Independent Variables

```python
x = new_ds.drop(columns="Class", axis=1)
x.head()
```

```python
y = new_ds["Class"]
y.head()
```

# Standardization Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss
```

```python
x_scaled = pd.DataFrame(ss.fit_transform(x), columns=x.columns)
```

```python
x_scaled.head()
```

# Splitting into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.1, random_state=10, stratify=y)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Building the model

We will be using a logistic regression model as it is a binary classification problem and efficiency is high in this model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

# Model evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

Our model has an accuracy of 96% which is great

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
tp = cm[0][0]
fn = cm[1][0]
fp = cm[0][1]
tn = cm[1][1]
```

```python
accuracy = (tp + tn) / (tp + tn + fp + fn)
accuracy
```

```python
precision = tp / (tp + fp)
precision
```

```python
recall = tp / (tp + fn)
recall
```

```python
f1_score = 2 * ( precision * recall ) / ( precision + recall )
f1_score
```

ROC AUC Curve-
-------------

```python
prob = model.predict_proba(x_test)[:, 1]
prob
```

```python
fpr, tpr, threshold = roc_curve(y_test, prob)
```

```python
fpr, tpr, threshold
```

```python
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

A precision of 1.0 (100%) is an excellent score, indicating that all the transactions identified as fraudulent were indeed fraudulent, with no false positives. 

A recall of 0.924 (92.4%) is also quite good, meaning the model correctly identified a large portion of the actual fraudulent transactions.

The F1 score, is 0.9607(96.07%), indicating a strong balance between precision and recall. An F1 score close to 1.0 is generally considered very good.


Hence our model is performing well.
